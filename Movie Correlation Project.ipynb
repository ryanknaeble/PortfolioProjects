```python
# Visualize the correlations for numeric features

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Look at Company for correlation
```

```python
# Numerize object/string by changing type to category and using cat.codes to assign each unique value in a column a number.
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Correlation matrix comparing all data fields

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
# Unstack the correlations to view more easily

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the corr_pairs 

sorted_pairs = corr_pairs.sort_values(ascending = False)

sorted_pairs
```

```python
# Show only the high correlations

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
# Votes and budget have the highest correlation to gross earnings

# Company and all non-numeric fields have low correlation
```
